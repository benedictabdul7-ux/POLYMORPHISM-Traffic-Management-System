```python
class TrafficDevice:
    def activate(self):
        print("Traffic device has been activated")
class Trafficlight(TrafficDevice):
    def activate(self):
        print("Traffic Light: The Green Light is ON")
class SpeedCamera(TrafficDevice):
    def activate(self):
        print("Speed Camera: Vehicles are been monitored")
class PedestrianSignal(TrafficDevice):
    def activate(self):
        print("Pedestrian signal: People are waiting")

class EmergencySiren(TrafficDevice):
    def activate(self):
        print("Emergency Siren: Has been turned ON! , vehicles are approaching")
l = Trafficlight()
S = SpeedCamera()
P = PedestrianSignal()
E = EmergencySiren()
Devices = [l, S, P , E]
for device in Devices:
    device.activate()
```
